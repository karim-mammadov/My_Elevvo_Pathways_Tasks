## Data preprocessing


```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = df_features_30_sec.copy()

X = df.drop(['filename', 'length', 'label'], axis=1)
y = df['label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

print("Shape of scaled features (X):", X_scaled.shape)
print("Shape of encoded target (y):", y_encoded.shape)
```

## Model selection and training (tabular)


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Instantiate and train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

print("Model training complete.")
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained model(s) using appropriate metrics for multi-class classification.


**Reasoning**:
Import the necessary metrics from sklearn.metrics, use the trained model to make predictions on the test set, and then calculate and print the accuracy score, classification report, and confusion matrix.



```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(report)

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Spectrogram generation


```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

audio_root_dir = os.path.join(path, 'Data', 'genres_original')

spectrogram_output_dir = '/kaggle/working/spectrograms'
os.makedirs(spectrogram_output_dir, exist_ok=True)

for genre_dir in os.listdir(audio_root_dir):
    genre_path = os.path.join(audio_root_dir, genre_dir)

    if os.path.isdir(genre_path):
        print(f"Processing genre: {genre_dir}")
        genre_output_dir = os.path.join(spectrogram_output_dir, genre_dir)
        os.makedirs(genre_output_dir, exist_ok=True)

        for audio_file in os.listdir(genre_path):
            if audio_file.endswith('.wav'):
                audio_path = os.path.join(genre_path, audio_file)

                try:
                    y, sr = librosa.load(audio_path)
                    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
                    S_dB = librosa.power_to_db(S, ref=np.max)
                    fig, ax = plt.subplots(figsize=(10, 4))
                    img = librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', ax=ax)

                    plt.axis('off')
                    plt.gca().set_position([0, 0, 1, 1])

                    output_filename = os.path.splitext(audio_file)[0] + '.png'
                    output_path = os.path.join(genre_output_dir, output_filename)

                    plt.savefig(output_path, bbox_inches='tight', pad_inches=0)
                    plt.close(fig)

                except Exception as e:
                    print(f"Error processing {audio_file}: {e}")


print("\nSpectrogram generation complete.")
```

## Data preprocessing for cnn


```python
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import os

spectrogram_dir = '/kaggle/working/spectrograms'

IMG_HEIGHT = 128
IMG_WIDTH = 128

images = []
labels = []

for genre_dir in os.listdir(spectrogram_dir):
    genre_path = os.path.join(spectrogram_dir, genre_dir)

    if os.path.isdir(genre_path):
        for image_file in os.listdir(genre_path):
            if image_file.endswith('.png'):
                image_path = os.path.join(genre_path, image_file)
                img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                img_resized = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
                img_normalized = img_resized / 255.0
                images.append(img_normalized)
                labels.append(genre_dir)

images = np.array(images)
labels = np.array(labels)
images = np.expand_dims(images, axis=-1)
X_train_img, X_test_img, y_train_img, y_test_img = train_test_split(images, labels, test_size=0.2, random_state=42, stratify=labels)

print("Image preprocessing complete.")
print("Shape of training images:", X_train_img.shape)
print("Shape of testing images:", X_test_img.shape)
print("Shape of training labels:", y_train_img.shape)
print("Shape of testing labels:", y_test_img.shape)
```

## Cnn model selection and training


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train_img)
y_test_encoded = encoder.transform(y_test_img)

y_train_one_hot = to_categorical(y_train_encoded)
y_test_one_hot = to_categorical(y_test_encoded)

num_classes = len(encoder.classes_)

model_cnn = Sequential()

model_cnn.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 1)))
model_cnn.add(MaxPooling2D(pool_size=(2, 2)))

model_cnn.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model_cnn.add(MaxPooling2D(pool_size=(2, 2)))

model_cnn.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model_cnn.add(MaxPooling2D(pool_size=(2, 2)))

model_cnn.add(Flatten())
model_cnn.add(Dense(128, activation='relu'))
model_cnn.add(Dense(num_classes, activation='softmax')) # Output layer with num_classes units and softmax activation

model_cnn.compile(loss='categorical_crossentropy',
                  optimizer=Adam(),
                  metrics=['accuracy'])

print("CNN model architecture defined and compiled.")
model_cnn.summary()
```

**Reasoning**:
Train the compiled CNN model using the preprocessed training data and one-hot encoded labels, specifying the number of epochs and batch size. Use a validation split to monitor performance during training.



```python
#CNN model
history = model_cnn.fit(X_train_img, y_train_one_hot,
                        epochs=20,
                        batch_size=32,
                        validation_split=0.2)

print("\nCNN model training complete.")
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

y_pred_cnn = model_cnn.predict(X_test_img)
y_pred_cnn_classes = np.argmax(y_pred_cnn, axis=1)

confusion_matrix_cnn = confusion_matrix(y_test_encoded, y_pred_cnn_classes)
print("Confusion Matrix (CNN):")
print(confusion_matrix_cnn)

plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix_cnn, annot=True, fmt='d', cmap='Blues', xticklabels=encoder.classes_, yticklabels=encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (CNN)')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
import os

data_path = os.path.join(path, 'Data')
features_30_sec_path = os.path.join(data_path, 'features_30_sec.csv')

if os.path.exists(features_30_sec_path):
    df_features_30_sec = pd.read_csv(features_30_sec_path)
    print("features_30_sec.csv loaded successfully.")
else:
    print(f"Error: {features_30_sec_path} not found.")

df = df_features_30_sec.copy()

X = df.drop(['filename', 'length', 'label'], axis=1)
y = df['label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

svm_model = SVC()
svm_model.fit(X_train, y_train)

print("SVM model training complete.")

y_pred_svm = svm_model.predict(X_test)

accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"Accuracy (SVM): {accuracy_svm:.4f}")

report_svm = classification_report(y_test, y_pred_svm)
print("\nClassification Report (SVM):")
print(report_svm)

conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
print("\nConfusion Matrix (SVM):")
print(conf_matrix_svm)

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix_svm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (SVM)')
plt.show()
```
