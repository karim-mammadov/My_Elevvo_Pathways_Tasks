# Customer Segmentation
## Description
The goal of this task is to **perform clustering analysis** on the Mall Customer dataset.  
Customers will be grouped based on their **Annual Income** and **Spending Score** to identify distinct segments and understand **customer behavior patterns**.

1. Load the Mall Customer dataset into a pandas DataFrame.

3. Preprocess data if necessary (e.g., scaling features).  
4. Apply clustering algorithms (e.g., K-Means, Hierarchical Clustering).  
5. Determine the optimal number of clusters using metrics like the **Elbow Method** or **Silhouette Score**.  
6. Analyze and visualize the resulting clusters to interpret customer segments.

### Data Loading and Initial Exploration


```python
import pandas as pd

# Assuming the dataset is named 'Mall_Customers.csv'
df = pd.read_csv('Mall_Customers.csv')
print("Mall Customer dataset loaded successfully.")
display(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

### Feature Selection


```python
if df is not None:
    X_clustering = df[['Annual Income (k$)', 'Spending Score (1-100)']]
    print("\nSelected features for clustering:")
    display(X_clustering.head())
else:
    print("Mall Customer DataFrame not loaded. Cannot select features.")
    X_clustering = None
```

### Data Scaling


```python
if X_clustering is not None:
    from sklearn.preprocessing import StandardScaler

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_clustering)
    print("\nScaled features:")
    print(X_scaled[:5])
else:
    print("Features not selected. Cannot perform scaling.")
    X_scaled = None
```

### Determine Optimal Number of Clusters (ELBOW METHOD)


```python
if X_scaled is not None:
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_score
    import matplotlib.pyplot as plt

    wcss = []
    silhouette_scores = []
    max_k = 10

    for i in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
        kmeans.fit(X_scaled)
        wcss.append(kmeans.inertia_)
        if i > 1:
            silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

    plt.figure(figsize=(10, 5))
    plt.plot(range(1, max_k + 1), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('WCSS')
    plt.xticks(range(1, max_k + 1))
    plt.grid(True)
    plt.show()

    if len(silhouette_scores) > 0:
        plt.figure(figsize=(10, 5))
        plt.plot(range(2, max_k + 1), silhouette_scores, marker='o', linestyle='--')
        plt.title('Silhouette Scores')
        plt.xlabel('Number of Clusters (K)')
        plt.ylabel('Silhouette Score')
        plt.xticks(range(2, max_k + 1))
        plt.grid(True)
        plt.show()

else:
    print("Scaled data not available. Cannot determine optimal number of clusters.")
```

### Apply K-Means Clustering


```python
if X_scaled is not None:

    optimal_k = 5

    kmeans_final = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)
    clusters = kmeans_final.fit_predict(X_scaled)


    if df is not None:
        df['Cluster'] = clusters
        print(f"\nK-Means clustering applied with K={optimal_k}.")
        display(df.head())
else:
    print("Scaled data not available. Cannot apply K-Means clustering.")
```

### Visualize Clusters

```python
if df is not None and 'Cluster' in df.columns:
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis', s=50)
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.title('Customer Clusters (K-Means)')
    plt.colorbar(scatter, label='Cluster')
    plt.grid(True)
    plt.show()
else:
    print("Clustered DataFrame not available. Cannot visualize clusters.")
```

### Bonus: Analyze average spending per cluster


```python
if df is not None and 'Cluster' in df.columns and 'Spending Score (1-100)' in df.columns:
    print("\nAverage Spending Score per K-Means Cluster:")
    average_spending_per_cluster = df.groupby('Cluster')['Spending Score (1-100)'].mean()
    display(average_spending_per_cluster)

    if 'Annual Income (k$)' in df.columns and 'Age' in df.columns:
         print("\nAverage Annual Income and Age per K-Means Cluster:")
         average_cluster_features = df.groupby('Cluster')[['Annual Income (k$)', 'Age']].mean()
         display(average_cluster_features)

else:
    print("Clustered DataFrame or required columns not available. Cannot analyze average spending per cluster.")
```
